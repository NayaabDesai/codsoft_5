```python
creditCard = creditCard.drop_duplicates()
```

```python
creditCard['Class'].value_counts()
```

```python
valid = creditCard[creditCard.Class == 0]
fraud = creditCard[creditCard.Class == 1]
```

```python
valid.Amount.describe()
```

```python
valid_sample = valid.sample(n=473)
valid_sample
```

```python
new_dataset = pd.concat([valid_sample, fraud], axis=0)
```

```python
new_dataset.head()
```

```python
#to find the relevant features the correlation of all the attributes has been derived
correlation = new_dataset.corr()
#correlation of all attributes with "Class" is derived in descending order to find the most relevant features 
corr_target = correlation['Class'].sort_values(ascending=False)
corr_target
```

```python
#hence 'V11','V4','V2','V21','V19','V20','V8','V27' are most relevant 
x = creditCard[['V11','V4','V2','V21','V19','V20','V8','V27']]
y = creditCard['Class']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

# LOGISTIC REGRESSION

```python
model = LogisticRegression()
model.fit(x_train,y_train)
```

```python
model_pred = model.predict(x_test)
```

```python
accuracy = accuracy_score(y_test,model_pred)
print("Logistic Regression Accuracy",accuracy*100)
```

```python
class_report = classification_report(y_test,model_pred)
print("Logistic Regression Classification_report:\n",class_report)
```

# DECISION TREE

```python
model2 = DecisionTreeClassifier(random_state=42)
model2.fit(x_train,y_train)
```

```python
model2_pred = model2.predict(x_test)
```

```python
accuracy2 = accuracy_score(y_test,model2_pred)
print("Decision Tree Accuracy:",accuracy2*100)
```

```python
class_report2 = classification_report(y_test,model2_pred)
print("Decision Tree Classification_report:\n",class_report2)
```
